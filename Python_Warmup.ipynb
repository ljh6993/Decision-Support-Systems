```python
def positionWord(filePath, position):
  # Please write your code here
  with open(filePath, "r") as textfile:
     text= textfile.read() # return string
  textlines=text.splitlines()
  list_words=[]
  for i in range(len(textlines)):
    textwords=textlines[i].split() # turn to word
    a = textwords[position].strip(""" ,.*()[]!@#$%^&*{}?'`"-""") #strip the symbol
    list_words.append(a)
  return(list_words)
    
    
```

```python
def onlyUpperCase(text):
  # Please write your code here
  str_1=''
  for i in text: 
    if i.isupper(): # if the letter is upper letter, return it 
      str_1=str_1+i
  return str_1

```

```python
def divisors(num1, num2):
  # Please write your code here
  list_1=[]
  for i in range(num1):
    if ((i+1)%num2)==0: # no remainder
      list_1.append(i+1)
  return list_1

```

```python
import pandas as pd
import numpy as np
url='https://raw.githubusercontent.com/MIE451-1513-2019/course-datasets/master/toronto_airbnb_listings_Aug2019.csv'
data = pd.read_csv(url,sep=",") # use sep="," for coma separation. 


def AvgNeighbourhoodListingPrice(neighbourhoodList, room_type, number_of_reviews, minimum_nights):
  # Please write your code here

    #Whether each element in the DataFrame is contained in values
    nbh=data[data['neighbourhood'].isin(neighbourhoodList)]
    rt=nbh[nbh['room_type']==room_type]
    nor=rt[rt['number_of_reviews']>=number_of_reviews]
    mn=nor[nor['minimum_nights']<=minimum_nights]

    #get the average price
    price=mn[['neighbourhood','price']].groupby(['neighbourhood']).mean().round(2)
    price=price.sort_values(by="price",ascending=False)
    return price
  
```

```python
import pandas as pd
import numpy as np

def leastSquaresFit(urlX,urlY):
  # Please write your code here
  x = pd.read_csv(urlX) # use sep="," for coma separation. 
  y= pd.read_csv(urlY)

  xnp=x.to_numpy()
  ynp=y.to_numpy()


  w=np.linalg.inv(xnp.T@xnp)@xnp.T@ynp # @ mutmal, *
  w=w[0:3]

  return w
```
